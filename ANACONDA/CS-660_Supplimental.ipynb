```python
def buildTestArray():
    """Builds a Test Array."""
    fullArray = []
    colList = []

    for row in range(4):
        for col in range(6):
            colList.append(float(col + (6 * row)))
            
        colList.reverse()
        
        for element in colList:
            fullArray.append(element)

        colList = []
    return fullArray
```

```python
def flopList( theList, rowLength=640):
    """Flop a depth array [python list]"""
    
    floppedList = []
    
    while( len(theList) > 0 ):

        oneRow = theList[:rowLength]
        theList = theList[rowLength:]

        oneRow.reverse()
        for element in oneRow:
            floppedList.append( element )
        
    
    return floppedList

```

Basically  NAN values are changed to 10000.00

```python
import os
import csv


def readCSVFile( file ):
    """
    Reads the CSV File.
    
    Parameters:
    File - File Name
    """
    returnList = []
    
    with open(file, newline='') as csvFile:
        spamReader = csv.reader(csvFile, delimiter=',')
        for row in spamReader:
            for element in row:
                returnList.append(element)
#             print(', '.join(row))

    return returnList


def processCSVData( data ):
    """
    Processes the CSV File.
    
    Parameters:
    Data - the Data as a list
    """
    returnList = []
    
    for element in data:
        if ( "nan" == element ):
            returnList.append("10000.00")
        else:
            returnList.append(element)

    return returnList


def writeCSVFile( data, file ):
    """
    Write the CSV File.
    
    Parameters:
    Data - the Data as a list
    File - the name of the file
    """
    with open(file, 'w', newline='') as csvFile:
        spamWriter = csv.writer(csvFile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
        spamWriter.writerow( data )
        
        
pCloudOrgNormDir = os.path.join( os.getcwd(), os.pardir, 'DATA', 'ORGANIZED', 'PCLOUD_NORM' )
pCloudOrgFlopDir = os.path.join( os.getcwd(), os.pardir, 'DATA', 'ORGANIZED', 'PCLOUD_FLOP' )

pCloudPrcNormDir = os.path.join( os.getcwd(), os.pardir, 'DATA', 'PROCESSED', 'PCLOUD_NORM' )
pCloudPrcFlopDir = os.path.join( os.getcwd(), os.pardir, 'DATA', 'PROCESSED', 'PCLOUD_FLOP' )

print( pCloudOrgNormDir )
print( pCloudOrgFlopDir )

onlyFilesInOrgNorm = [f for f in os.listdir(pCloudOrgNormDir) if os.path.isfile(os.path.join(pCloudOrgNormDir, f))]
onlyFilesInOrgFlop = [f for f in os.listdir(pCloudOrgFlopDir) if os.path.isfile(os.path.join(pCloudOrgFlopDir, f))]


onlyFilesInOrgNorm = onlyFilesInOrgNorm[1:]
onlyFilesInOrgFlop = onlyFilesInOrgFlop[1:]

# print(onlyFilesInOrgNorm)

numberOfFiles = 0
maxFloat = 0.0

for element in onlyFilesInOrgNorm:
#     if (element == 'TEST2.csv'):
    testFile = os.path.join( pCloudOrgNormDir, element )
    testWriteFile = os.path.join( pCloudPrcNormDir, element )

    data = readCSVFile( testFile )
    data = processCSVData( data )

#     for datum in data:
#         if ( float(datum) > maxFloat ):
#             maxFloat = float(datum)

    writeCSVFile( data, testWriteFile )
    
    numberOfFiles = numberOfFiles + 1
    print( element )
    print( str(numberOfFiles) + " FILES WRITTEN")
        
# print( "" )
# print( "MAX FLOAT: " + str(maxFloat) )
# print( "" )

numberOfFiles = 0        

for element in onlyFilesInOrgFlop:
#     if (element == 'TEST2.csv'):
    testFile = os.path.join( pCloudOrgFlopDir, element )
    testWriteFile = os.path.join( pCloudPrcFlopDir, element )

    data = readCSVFile( testFile )
    data = processCSVData( data )
    writeCSVFile( data, testWriteFile )

    numberOfFiles = numberOfFiles + 1
    print( element )
    print( str(numberOfFiles) + " FILES WRITTEN")

        
```
